```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Graph

```python
# defining the x axis values and y axis values
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize graph
plt.figure(figsize=(5,3),dpi=100)

## line 1

# Keyword argument Notation
# plt.plot(x,y, label ='2x',color='RED',linewidth=2,marker=".",markersize=10,markeredgecolor='BLUE',linestyle='--')

#Shor hand the Notation
# after the variables 'r.--' = [color][marker][line]
plt.plot(x,y,'b^--' ,label ='2x')

## line 2

#select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

#plot part of the graph as line 
plt.plot(x2[:6],x2[:6]**2,'r',label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:],x2[5:]**2,'r--')

#Add a title with the fontdict to specify font details
plt.title("Our First Graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})

#x,y axis labels
plt.xlabel("X Axis")#,fontdict={'fontname':'Arial'})
plt.ylabel("Y Axis")

# X, Y axis tickmarks(scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

#add legend
plt.legend()

#save figure
plt.savefig('mygraph.png',dpi=300)

# show plot of the graph
plt.show()
```

## Real World Examples
##### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time  (in USD)',fontdict={'fontweight':'bold','fontsize':18})

plt.plot(gas.Year,gas.USA,'r.-', label='United States')
plt.plot(gas.Year,gas.Canada,'b.-' ,label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-', label='South Korea')
plt.plot(gas.Year,gas.Australia,'y.-' ,label='Australia')

#another way to plot many values::

# countries = ['Australia','USA','Canada','South Korea']
# for country in gas:
#     if country in countries:
#         plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.jpeg',dpi=1000)

plt.show()
```

### FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

#### Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize =(8,5),dpi=75)

plt.hist(fifa.Overall,bins=bins)

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skill in Fifa 2018',fontdict={'fontweight':'bold','fontsize':18})

plt.show()
```

#### Pie Charts

```python
plt.figure(figsize=(8,5),dpi=100)

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['green','skyblue']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f %%')

plt.title('Foot Preference of FIFA players',fontdict={'fontweight':'bold','color':'red','fontsize':18})

plt.show()
```

### Pie Chart 2

```python
plt.figure(figsize=(8,5),dpi=100)
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
labels = ['Under 125','125-150','150-175','175-200', 'Over 200']

explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)',fontdict={'fontweight':'bold','fontsize':18})
plt.pie(weights,labels=labels, autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})
 
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
